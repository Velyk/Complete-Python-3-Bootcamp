```python
from IPython.display import clear_output

def display_board(board):
    
    clear_output()
    print(" | | " )
    print(board[7] + "|" + board[8] + "|" + board[9])
    print("_|_|_" )
    print(" | | " )
    print(board[4] + "|" + board[5] + "|" + board[6])
    print("_|_|_" )
    print(" | | " )
    print(board[1] + "|" + board[2] + "|" + board[3])
    print(" | | " )
```

```python
def player_input():
    
    choice = "Wrong"
    
    while choice not in ["X","O","x","o"]:
        
        choice = input("Player 1 please select X or O: ")
           
    player1 = choice.upper()
    
    if player1 == "X":
        player2 = "O"
    
    else:
        player2 = "X"
        
    return (player1,player2)
```

```python
## DETERMINE PLAYER ORDER

import random

def choose_first():
    
    first = random.randint(1, 2)
    
    print(f"player {first} goes first")
    
    return first
```

```python
from IPython.display import clear_output

def place_marker(board, marker, position):

    if position not in range(1,10):
            print("Please select a valid position from 1 to 9!")
            
    if board[position] == " ":
        
        board[position] = marker
    else:
        print("choose another space")
        
        
```

```python
def full_board_check(board):
    
    for i in range(1,9):
        if board[i] == " ":
            return False
        else:
            return True
```

```python
def player_choice(board):
    
        position = "Wrong"
        
        while position not in range(1,10):
        
            position = int(input("Please select a board position from 1 to 9: "))
        
        if board[position] == " ":
            return position
        else:
            print("choose another space")
            return position
```

```python
def win_check(board, mark):
    
    if board[1] == mark and board[2] == mark and board[3] == mark:
        return True
    elif board[4] == mark and board[5] == mark and board[6] == mark:
        return True
    elif board[7] == mark and board[8] == mark and board[9] == mark:
        return True
    elif board[1] == mark and board[5] == mark and board[9] == mark:
        return True
    elif board[8] == mark and board[5] == mark and board[2] == mark:
        return True
    elif board[7] == mark and board[5] == mark and board[3] == mark:
        return True
    else:
        return False
```

```python
def player1_turn():
    
    if first_turn == 1:
        
        #Player1
        position = player_choice(game_board)
        
        place_marker(game_board, p1 , position)
        
        display_board(game_board)
        
        if win_check(game_board,p1) == True:
        
            print("Player 1 Wins!!!")
            
            game_board = ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']
        
        #Player 2
        
        position = player_choice(game_board)
        
        place_marker(game_board, p2 , position)
        
        display_board(game_board)
        
        if win_check(game_board,p2) == True:
        
            print("Player 2 Wins!!!")
            
            game_board = ['#',' ',' ',' ',' ',' ',' ',' ',' ',' ']
```
